```python

class OptimalPolicyAgent:
    def __init__(self, env: TaskEnv_driftype):
        self.policy = None
        self.state_value =None
        actionlist = env.motions
        self.actions = {a: i for i, a in enumerate(actionlist)}
        
    def act(self, state):
        actionname = self.policy[state]
        return self.actions[actionname]

    def value_iteration(self, gamma=0.99, theta=1e-6):
        state_list = env.states
        action_list = env.motions
        state_idx = {s: i for i, s in enumerate(state_list)}
        
        n_states = len(state_list)

        V = np.zeros(n_states)
        policy = np.zeros(n_states, dtype=int)

        while True:
            delta = 0
            for s_name in state_list :
                if s_name =='Tau':
                    continue
                s = state_idx[s_name]
                q_values = []
                for a_name in action_list:
                    q = 0
                    for s_prime_name, prob in env.observation_space[s_name][a_name].items():
                        s_prime = state_idx[s_prime_name]
                        reward = env.severity.get(s_prime_name, 0)
                        if s_prime == 'Tau':
                            q+= prob*reward
                        else:
                            q += prob * (reward + gamma * V[s_prime])
                    q_values.append(q)

                max_q = max(q_values)
                delta = max(delta, abs(V[s] - max_q))
                V[s] = max_q
                policy[s] = np.argmax(q_values)
                
            if delta < theta:
                break

        optimal_policy = {state_list[s]: action_list[policy[s]] for s in range(n_states) if state_list[s]!='Tau'}

        self.policy = optimal_policy
        self.state_value = V


def evaluate_agent_opt(env, agent, episodes=1000,drift =400):
    total_rewards = []
    agent.value_iteration()
    for e in range(episodes):
        if e==drift:
            env.set_flag()
            env.drift(add_actions, change_at_states, drift_dis_type,intensity, drift_type)
            agent.value_iteration()

        for _ in range(episodes):
            state = env.reset()
            done = False
            episode_reward = 0
            while not done:
                action = agent.act(state)
                state, reward, done, _ = env.step(action)
                episode_reward += reward
        total_rewards.append(episode_reward)
    return total_rewards

```

```python
class OptimalPolicyAgent:
    def __init__(self, env: TaskEnv_driftype):
        self.env = env
        self.policy = None
        self.state_value = None
        self.action_list = env.motions
        self.state_list = env.states
        self.state_idx = {s: i for i, s in enumerate(self.state_list)}
        self.actions = {a: i for i, a in enumerate(self.action_list)}

    def act(self, state):
        actionname = self.policy[state]
        return self.actions[actionname]

    def policy_iteration(self, gamma=0.95, theta=1e-6):
        n_states = len(self.state_list)
        n_actions = len(self.action_list)
        V = np.zeros(n_states)
        policy_idx = np.random.randint(n_actions, size=n_states)  # 初始随机策略

        while True:
            # === Policy Evaluation ===
            while True:
                delta = 0
                for s_name in self.state_list:
                    if s_name == 'Tau':
                        continue
                    s = self.state_idx[s_name]
                    a_name = self.action_list[policy_idx[s]]
                    v = V[s]
                    new_v = 0
                    for s_prime_name, prob in self.env.observation_space[s_name][a_name].items():
                        s_prime = self.state_idx[s_prime_name]
                        reward = self.env.severity.get(s_prime_name, 0)
                        if s_prime_name == 'Tau':
                            new_v += prob * reward
                        else:
                            new_v += prob * (reward + gamma * V[s_prime])
                    delta = max(delta, abs(v - new_v))
                    V[s] = new_v
                if delta < theta:
                    break

            # === Policy Improvement ===
            policy_stable = True
            for s_name in self.state_list:
                if s_name == 'Tau':
                    continue
                s = self.state_idx[s_name]
                old_action = policy_idx[s]

                q_values = []
                for a_name in self.action_list:
                    q = 0
                    for s_prime_name, prob in self.env.observation_space[s_name][a_name].items():
                        s_prime = self.state_idx[s_prime_name]
                        reward = self.env.severity.get(s_prime_name, 0) - self.get_action_penalty(a_name)
                        if s_prime_name == 'Tau':
                            q += prob * reward
                        else:
                            q += prob * (reward + gamma * V[s_prime])
                    q_values.append(q)

                best_action = int(np.argmax(q_values))
                policy_idx[s] = best_action
                if best_action != old_action:
                    policy_stable = False

            if policy_stable:
                break

        # 保存策略和值函数
        self.state_value = V
        self.policy = {self.state_list[s]: self.action_list[policy_idx[s]] for s in range(n_states) if self.state_list[s] != 'Tau'}
```

```python
import matplotlib.pyplot as plt
import numpy as np

def evaluate_agent(env, agent, n_episodes=100, max_steps=50):
    episodic_rewards = []

    for _ in range(n_episodes):
        state = env.reset()
        total_reward = 0

        for _ in range(max_steps):
            action = agent.act(state)
            next_state, reward, done, _ = env.step(action)
            total_reward += reward
            if done or next_state == 'Tau':
                break
            state = next_state

        episodic_rewards.append(total_reward)

    return episodic_rewards
```
